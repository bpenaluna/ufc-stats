### Preprocessing

```python
print('Unique red fighters: {}\nUnique blue fighters: {}'.format(df['RedFighter'].nunique(), df['BlueFighter'].nunique()))
```

```python
fighters_list = np.unique(np.concatenate((df['RedFighter'], df['BlueFighter'])))
print('Number of unique fighters:', len(np.unique(fighters_list)))
```

```python
fights_long = pd.DataFrame({
    'Fighter': pd.concat([df['RedFighter'], df['BlueFighter']], ignore_index=True),
    'FinishRound': pd.concat([df['FinishRound'], df['FinishRound']], ignore_index=True),
})

fights_long.head()
```

```python
fights_long.shape
```

```python
# fighter_stats = fights_long.groupby('Fighter').agg({
#     'Wins': 'mean',
# })

# fighter_stats.columns = ['AvgWins']  # rename aggregated columns
# fighter_stats.head()
```

```python
# 'Red' -> 1, 'Blue' -> 0 for Winner column
df['Winner'].replace({'Red': 1, 'Blue': 0}, inplace=True)
```

```python
# True -> 1, False -> 0 for TitleBout column
df['TitleBout'].replace({True: 1, False: 0}, inplace=True)
```

```python
# Removes 'Women's' from WeightClass column
df.loc[df['Gender'] == 0, 'WeightClass'] = df.loc[df['Gender'] == 0, 'WeightClass'].str.split().str[1]
```

```python
# Convert string Date column to datetime type
# Engineers 'DaysSinceFight' column with number of days since the fight date and removes Date column
date_format = '%Y-%m-%d'
df['DaysSinceFight'] = df['Date'].map(lambda x: (datetime.now() - datetime.strptime(x, date_format)).days)
```

```python
df['Date'].map(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
def convert_class_rank_str(gender, weight):
    g = 'W' if gender == 'FEMALE' else ''
    return g + weight + 'Rank'
```

```python
df['BRank'] = df.apply(
    lambda x: x.get('B' + convert_class_rank_str(x['Gender'], x['WeightClass'])), axis=1
)
df['RRank'] = df.apply(
    lambda x: x.get('R' + convert_class_rank_str(x['Gender'], x['WeightClass'])), axis=1
)

df.fillna({'BRank': 16, 'RRank': 16}, inplace=True)
```

```python
df['Gender'].replace({'MALE': 1, 'FEMALE': 0}, inplace=True)
```

```python
df['RPFP'] = [1 if type(x) == float else 0 for x in df['RPFPRank']]
df['BPFP'] = [1 if type(x) == float else 0 for x in df['BPFPRank']]
```

```python
weight_rank_cols = [
    'BMatchWCRank',
    'RMatchWCRank',
    'RWFlyweightRank',
    'RWFeatherweightRank',
    'RWStrawweightRank',
    'RWBantamweightRank',
    'RHeavyweightRank',
    'RLightHeavyweightRank',
    'RMiddleweightRank',
    'RWelterweightRank',
    'RLightweightRank',
    'RFeatherweightRank',
    'RBantamweightRank',
    'RFlyweightRank',
    'RPFPRank',
    'BWFlyweightRank',
    'BWFeatherweightRank',
    'BWStrawweightRank',
    'BWBantamweightRank',
    'BHeavyweightRank',
    'BLightHeavyweightRank',
    'BMiddleweightRank',
    'BWelterweightRank',
    'BLightweightRank',
    'BFeatherweightRank',
    'BBantamweightRank',
    'BFlyweightRank',
    'BPFPRank'
]
df.drop(columns=weight_rank_cols, inplace=True)
```

```python
def min_to_sec(mins):
    arr = mins.split(":")
    return int(arr[0]) * 60 + int(arr[1])
```

```python
df.fillna({'FinishRoundTime': '5:00'}, inplace=True)
df['FinishRoundTime'] = df['FinishRoundTime'].map(lambda x: min_to_sec(str(x)))
```

```python
sns.violinplot(data=df, x='BlueExpectedValue')
```

```python
df['Rankdif'] = abs(df['BRank'] - df['RRank'])
df['OddsDif'] = df['BlueOdds'] - df['RedOdds']
df['KOOddsDif'] = df['BKOOdds'] - df['RKOOdds']
df['SubOddsDif'] = df['BSubOdds'] - df['RSubOdds']
df['DecOddsDif'] = df['BlueDecOdds'] - df['RedDecOdds']
df['AvgSigStrLandedDif'] = df['BlueAvgSigStrLanded'] - df['RedAvgSigStrLanded']
df['AvgTDPctDif'] = df['BlueAvgTDPct'] - df['RedAvgTDPct']
df['AvgTDLandedDif'] = df['BlueAvgTDLanded'] - df['RedAvgTDLanded']
df['AvgSubAttDif'] = df['BlueAvgSubAtt'] - df['RedAvgSubAtt']
df['AvgSigStrPctDif'] = df['BlueAvgSigStrPct'] - df['RedAvgSigStrPct']
df['CurrentWinStreakDif'] = df['BlueCurrentWinStreak'] - df['RedCurrentWinStreak']
df['CurrentLoseStreakDif'] = df['BlueCurrentLoseStreak'] - df['RedCurrentLoseStreak']
df['DrawsDif'] = df['BlueDraws'] - df['RedDraws']
df['WinsDif'] = df['BlueWins'] - df['RedWins']
df['LossesDif'] = df['BlueLosses'] - df['RedLosses']
df['WeightLbsDif'] = df['BlueWeightLbs'] - df['RedWeightLbs']
df['WinsByMDecDif'] = df['BlueWinsByDecisionMajority'] - df['RedWinsByDecisionMajority']
df['WinsBySDecDif'] = df['BlueWinsByDecisionSplit'] - df['RedWinsByDecisionSplit']
df['WinsByUDecDif'] = df['BlueWinsByDecisionUnanimous'] - df['RedWinsByDecisionUnanimous']
df['WinsByKODif'] = df['BlueWinsByKO'] - df['RedWinsByKO']
df['WinsBySubDif'] = df['BlueWinsBySubmission'] - df['RedWinsBySubmission']
df['WinsByTKODif'] = df['BlueWinsByTKODoctorStoppage'] - df['RedWinsByTKODoctorStoppage']
df['ExpectedValueDif'] = df['BlueExpectedValue'] - df['RedExpectedValue']
df['TotalRoundsFoughtDif'] = df['BlueTotalRoundsFought'] - df['RedTotalRoundsFought']
df['TotalTitleBoutsDif'] = df['BlueTotalTitleBouts'] - df['RedTotalTitleBouts']
df['HeightDif'] = df['BlueHeightCms'] - df['RedHeightCms']
df['ReachDif'] = df['BlueReachCms'] - df['RedReachCms']

df.drop(
    columns=[
        'Date',
        'BlueAge', 'RedAge',
        'BRank', 'RRank',
        'BlueOdds', 'RedOdds',
        'BKOOdds', 'RKOOdds',
        'BSubOdds', 'RSubOdds',
        'BlueDecOdds', 'RedDecOdds',
        'BlueAvgSigStrLanded', 'RedAvgSigStrLanded',
        'BlueAvgTDPct', 'RedAvgTDPct',
        'BlueAvgTDLanded', 'RedAvgTDLanded',
        'BlueAvgSubAtt', 'RedAvgSubAtt',
        'BlueAvgSigStrPct', 'RedAvgSigStrPct',
        'BlueCurrentWinStreak', 'RedCurrentWinStreak',
        'BlueCurrentLoseStreak', 'RedCurrentLoseStreak',
        'BlueDraws', 'RedDraws',
        'BlueWins', 'RedWins',
        'BlueLosses', 'RedLosses',
        'BlueWeightLbs', 'RedWeightLbs',
        'BlueWinsByDecisionMajority', 'RedWinsByDecisionMajority',
        'BlueWinsByDecisionSplit', 'RedWinsByDecisionSplit',
        'BlueWinsByDecisionUnanimous', 'RedWinsByDecisionUnanimous',
        'BlueWinsByKO', 'RedWinsByKO',
        'BlueWinsBySubmission', 'RedWinsBySubmission',
        'BlueWinsByTKODoctorStoppage', 'RedWinsByTKODoctorStoppage',
        'BlueExpectedValue', 'RedExpectedValue',
        'BlueLongestWinStreak', 'RedLongestWinStreak',
        'BlueTotalRoundsFought', 'RedTotalRoundsFought',
        'BlueTotalTitleBouts', 'RedTotalTitleBouts',
        'BlueHeightCms', 'RedHeightCms',
        'BlueReachCms', 'RedReachCms',
    ],
    inplace=True
)
```

```python
# Convert odds: American -> Probabilistic
df['OddsDif'] = [100 / (x + 100) if x > 0 else (-x + 100) / 100 for x in df['OddsDif']]
df['KOOddsDif'] = [100 / (x + 100) if x > 0 else (-x + 100) / 100 for x in df['KOOddsDif']]
df['SubOddsDif'] = [100 / (x + 100) if x > 0 else (-x + 100) / 100 for x in df['SubOddsDif']]
df['DecOddsDif'] = [100 / (x + 100) if x > 0 else (-x + 100) / 100 for x in df['DecOddsDif']]
```

```python
list(df.columns.sort_values())
```

```python
df.fillna(
    {
        'KOOddsDif': 0,
        'SubOddsDif': 0,
        'DecOddsDif': 0,
        'AvgSigStrLandedDif': 0,
        'AvgTDPctDif': 0,
        'AvgTDLandedDif': 0,
        'AvgSubAttDif': 0,
        'AvgSigStrPctDif': 0,
        'ExpectedValueDif': 0,
        'OddsDif': 0,
        'TotalFightTimeSecs': 3 * 5 * 60,
        'FinishRound': 3,
        'EmptyArena': 0.0,
        # 'BlueStance': df['BlueStance'].mode()[0]
    },
    inplace=True
)
```

```python
decay_lambda = 0.001
```

```python
# df['DecayedBRank'] = np.exp(-1 * df['BRank'] * df['DaysSinceFight'] * decay_lambda)
# df['DecayedBRank']
```

```python
non_decay_cols = [
    'RedFighter',
    'BlueFighter',
    'Location',
    'Country',
    'Winner',
    'TitleBout',
    'WeightClass',
    'NumberOfRounds',
    'BlueStance',
    'RedStance',
    'HeightDif',
    'ReachDif',
    'WeightLbsDif',
    'BetterRank',
    'Finish',
    'FinishDetails',
    'DaysSinceFight',
    'RPFP',
    'BPFP',
    'EmptyArena',
    'Gender'
]

decay_cols = df.drop(columns=non_decay_cols)
cat_col = [col for col in decay_cols.columns if decay_cols[col].dtypes == 'object']
cat_col
```

```python
for col in list(decay_cols.columns):
    df['Decayed' + col] = np.clip(np.exp(-1 * decay_cols[col] * df['DaysSinceFight'] * decay_lambda), -100000, 100000)
df.drop(columns=decay_cols, inplace=True)
```

```python
X, y = df.drop(columns='Winner'), df['Winner']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Label Encoding
label_enc_columns = ['RedFighter', 'BlueFighter']
enc = LabelEncoder()
enc.fit(fighters_list)
X_train['RedFighter'] = enc.transform(X_train['RedFighter'])
X_train['BlueFighter'] = enc.transform(X_train['BlueFighter'])
X_test['RedFighter'] = enc.transform(X_test['RedFighter'])
X_test['BlueFighter'] = enc.transform(X_test['BlueFighter'])
```

```python
# One Hot Encoding
ohe_columns = ['WeightClass', 'Finish', 'FinishDetails', 'BetterRank']
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
X_train = pd.concat([X_train, ohe.fit_transform(X_train[ohe_columns])], axis=1)
X_train.drop(columns=ohe_columns, inplace=True)
```

```python
X_test = pd.concat([X_test, ohe.transform(X_test[ohe_columns])], axis=1)
X_test.drop(columns=ohe_columns, inplace=True)
```

### Hyperparameter Tuning

```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
def rf_cl_bo(max_depth, max_features, n_estimators):
    params_rf = {}
    params_rf['max_depth'] = round(max_depth)
    params_rf['max_features'] = max_features
    params_rf['n_estimators'] = round(n_estimators)
    scores = cross_val_score(RandomForestClassifier(random_state=42, **params_rf),
                             X_train, y_train, scoring='accuracy', cv=5).mean()
    score = scores.mean()
    return score
```

```python
params_rf = {
    'max_depth': (3, 15),
    'max_features': (0.8, 1),
    'n_estimators': (50, 500)
}
```

```python
rf_bo = BayesianOptimization(rf_cl_bo, params_rf, random_state=42)
```
